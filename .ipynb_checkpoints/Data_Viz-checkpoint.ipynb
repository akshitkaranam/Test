```python
# Extract years
years = [str(year) for year in range(1980, 2029)]

# Create a DataFrame
df = countries_gdp
df = df.melt(id_vars=['Country'], var_name='Year', value_name='GDP Growth Rate')

# Create a line chart using Matplotlib
plt.figure(figsize=(12, 6))


i = 1
for country in df['Country'].unique():
    country_data = df[df['Country'] == country]
    
    x = country_data['Year'].astype(int)
    y = country_data['GDP Growth Rate']
    
#     # Set line style to dotted for years after 2022
#     if x.max() > 2022:
#         linestyle = 'dotted'
#     else:
#         linestyle = 'solid'

    if i== 1:
        color = "#1f77b4"
    else:
        color = '#2ca02c'
    
    plt.plot(x, y, label=country,color=color)
    i+=1


plt.xlabel('Year')
plt.ylabel('GDP Growth Rate')
plt.title('GDP Growth Rate Over Time')
plt.legend(loc='upper left') 

# Show the chart
plt.grid(False)
plt.savefig('gdp_growth_comparison_developing.png',transparent=True)

plt.show()
```
